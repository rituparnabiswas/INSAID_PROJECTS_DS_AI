Graphical Plotting:
Name of the FastFoodRestaurants in US on the basis of count

```python
from collections import Counter
name_count = Counter(ffd['name'].dropna().tolist()).most_common(10) #although null value not there but for standard coding maintained that
name_idx = [name[0] for name in name_count]
name_val = [name[1] for name in name_count]
fig,ax = plt.subplots(figsize=(8,6))
sns.barplot(x = name_idx,y=name_val ,ax =ax)
plt.title('Top ten name of restaurants in US')
plt.xlabel('Restaurants Name')
plt.ylabel('Count')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```

Observations:

So as we predicted earlier McDonals's is the highest in the number of presence in US.(>1750)
Now from the graph we can say BurgerKing is the second(>1000 and <1250) and Taco Bell is the third(>750 and <1000)

Plotting of provinces where maximum FastFoodRestaurants are present:

```python
from collections import Counter
province_count = Counter(ffd['province'].dropna().tolist()).most_common(10)
province_idx = [province[0] for province in province_count]
province_val = [province[1] for province in province_count]
fig,ax = plt.subplots(figsize=(8,6))
sns.barplot(x = province_idx,y=province_val ,ax =ax)
plt.title('Top ten provinces')
plt.xlabel('provinces')
plt.ylabel('Count')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```

Observations:

So from graphical representation we can say CA is in the top followed by TX and OH respectively.

Plotting of cities where maximum FastFoodRestaurants are present:

```python
from collections import Counter
city_count = Counter(ffd['city'].dropna().tolist()).most_common(10)
city_idx = [city[0] for city in city_count]
city_val = [city[1] for city in city_count]
fig,ax = plt.subplots(figsize=(8,6))
sns.barplot(x = city_idx,y=city_val ,ax =ax)
plt.title('Top ten cities')
plt.xlabel('province')
plt.ylabel('Count')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```

Observations:

So from graphical representation we can say Cincinnati is in the top followed by Las Vegas and Houston respectively.

Corelation between province and Latitude and Longitude:

The geographical location of an area has high impact in the case business.
Normally business area is populated and busy area. So there may be some corelation between the province and the geographical location of that province.

Below we are trying plot that corelation:

```python
df = pd.DataFrame({'province': ffd['province'], 'latitude': ffd['latitude'], 'longitude':ffd['longitude']}, dtype='category')         # Create a new dataframe with the three columns and assign numbers in place of their categories
df_num = df.apply(lambda x: x.cat.codes)

print("Pearson:", df_num.corr())                                                                                    # Run a correlation calculation 
print("")
```

```python
features = df_num.corr()
plt.figure(figsize=(4,4))
sns.heatmap(features,vmax=1,square=True,annot=False,cmap='Blues')
plt.show()
```

So from above reprensentation we can say there are corelation between province and latitude.

Now we will calculate and draw a graphical representation between latitude and province(CA,TX,OH and FL- FastFood restaurants countwise first four in US)

```python
ffd['latitude_group'] = pd.cut(ffd['latitude'].dropna(),    #creating a new column lalitude group based on the latitude range
                         [0,18,25,35,45,99],
                         labels=['<0','18-24','25-34','35-44','45+'], include_lowest=True)
```

```python
myprovince=['CA','TX','OH','FL']

ffd_selectedprovinceds=ffd[ffd['province'].isin(myprovince)]
```

```python
ffd_selectedprovinceds
```

Using Seaborn swarm Plot

```python
sns.set(style="whitegrid", palette="muted")


# "Melt" the dataset to "long-form" or "tidy" representation
pd.melt(ffd_selectedprovinceds, "province")

# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="province", y="latitude_group", 
              palette='gnuplot', data=ffd_selectedprovinceds)
```
